```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame(hist_data)

# Create histograms for each sample type
sample_types = df['Sample'].unique()

# Number basis histogram on probability scale
plt.figure(figsize=(10, 6))
for sample in sample_types:
    subset = df[df['Sample'] == sample]
    plt.hist(subset['FFT_#'], bins=np.arange(2,17), density=True, alpha=0.5, label=f'{sample} (Number)')
plt.xlabel(r'$X_R$', fontsize=16)
plt.ylabel('Probability', fontsize=16)
plt.xticks(np.arange(2, 18, 2))
plt.legend()
plt.title('Histogram by Number Basis (Probability Scale)', fontsize=18)
plt.show()

# Volume basis histogram on probability scale
plt.figure(figsize=(10, 6))
for sample in sample_types:
    subset = df[df['Sample'] == sample]
    plt.hist(subset['FFT_#'], bins=np.arange(2,17), density=True, alpha=0.5, weights=subset['Vol_wt'], label=f'{sample} (Volume)')
plt.xlabel(r'$X_R$', fontsize=16)
plt.ylabel('Probability', fontsize=16)
plt.xticks(np.arange(2, 18, 2))
plt.legend()
plt.title('Histogram by Volume Basis (Probability Scale)', fontsize=18)
plt.show()
```

```python
# Define custom colors
salmon_color = (0.7, 0.2, 0.2)  # Salmon tone in normalized RGB
blue_color = (0.3, 0.4, 0.8)  # A custom blue tone in normalized RGB

# Plot Number basis histogram on probability scale
plt.figure(figsize=(10, 6))
for sample in sample_types:
    subset = df[df['Sample'] == sample]
    if sample == 'CMP':
        plt.hist(subset['FFT_#'], bins=np.arange(2,17), density=True, alpha=0.5, color=salmon_color, label=f'{sample}')
    elif sample == 'GA':
        plt.hist(subset['FFT_#'], bins=np.arange(2,17), density=True, alpha=0.5, color=blue_color, label=f'{sample}')
    else:
        plt.hist(subset['FFT_#'], bins=np.arange(2,17), density=True, alpha=0.5, label=f'{sample}')

plt.xlabel(r'$X_R$', fontsize=16)
plt.ylabel('Probability', fontsize=16)
plt.xticks(np.arange(2, 18, 2))
plt.legend()
plt.title('Histogram by Number Basis (Probability Scale)', fontsize=18)
plt.show()

# Plot Volume basis histogram on probability scale
plt.figure(figsize=(10, 6))
for sample in sample_types:
    subset = df[df['Sample'] == sample]
    if sample == 'CMP':
        plt.hist(subset['FFT_#'], bins=np.arange(2,17), density=True, alpha=0.5, weights=subset['Vol_wt'], color=salmon_color, label=f'{sample}')
    elif sample == 'GA':
        plt.hist(subset['FFT_#'], bins=np.arange(2,17), density=True, alpha=0.5, weights=subset['Vol_wt'], color=blue_color, label=f'{sample}')
    else:
        plt.hist(subset['FFT_#'], bins=np.arange(2,17), density=True, alpha=0.5, weights=subset['Vol_wt'], label=f'{sample}')

plt.xlabel(r'$X_R$', fontsize=16)
plt.ylabel('Probability', fontsize=16)
plt.xticks(np.arange(2, 18, 2))
plt.legend()
plt.title('Histogram by Volume Basis (Probability Scale)', fontsize=18)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define a custom color palette
custom_palette = {
    "GA": "mediumblue",
    "CMP": "darkred"
}

# Sort data to ensure CMP is plotted last
df_sorted = df.sort_values(by="Sample", ascending=False)  # CMP on top

# Create a KDE plot without filling, but we'll handle transparency later
g = sns.jointplot(
    x="k_rsd",
    y="k_skew",
    hue="Sample",
    data=df_sorted,
    kind="kde",
    levels = 5,
    fill=False,   # Fill contours to distinguish densities
    alpha=0.8,   # General transparency
    palette=custom_palette,
    marginal_kws=dict(common_norm=False, fill=True)  # Marginal densities
)

# Adjust the alpha for each hue separately
for collection, sample in zip(g.ax_joint.collections, df_sorted["Sample"].unique()):
    if sample == "GA":  # More transparent for GA
        collection.set_alpha(0.2)
    else:  # Slightly opaque for CMP
        collection.set_alpha(0.5)

# Set axis labels and limits
g.ax_joint.set_xlim(-1, 6)
g.ax_joint.set_ylim(-6, 6)
g.ax_joint.set_yticks(np.arange(-6, 7, 2))
g.ax_joint.set_xticks(np.arange(0, 7, 2))
g.ax_joint.set_ylabel(r'$\kappa_{skew}$', fontsize=16)
g.ax_joint.set_xlabel(r'$\kappa_{RSD}$', fontsize=16)

plt.show()
```
